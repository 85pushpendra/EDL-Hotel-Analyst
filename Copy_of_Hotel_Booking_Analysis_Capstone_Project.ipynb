## **Data Cleaning and handaling**

**Step1**: Remove all the duplicate column in the data set

```python
# Check the duplicasy
new_df.duplicated()
```

```python
new_df[new_df.duplicated()].shape
```

```python
new_df.drop_duplicates(inplace = True)
new_df.shape
```

```python
# replace all the nan value with boolean and then sort
new_df.isnull().sum().sort_values(ascending = False)[:10]
```

```python
new_df[['company','agent']] = new_df[['company','agent']].fillna(0)
new_df[['company','agent']]
```

```python
new_df['children'].fillna(new_df['children'].mean(), inplace = True)
new_df['country'].fillna('others', inplace = True)
```

```python
new_df.isnull().sum().sort_values(ascending = False)[:10]
```

```python
new_df[new_df['adults']+new_df['babies']+new_df['children'] == 0].shape
```

```python
new_df.drop(new_df[new_df['adults']+new_df['children']+new_df['babies'] == 0].index, inplace = True)
```

# **Updat data Type** 

```python
new_df.info()
```

```python
new_df['reservation_status_date'] = pd.to_datetime(new_df['reservation_status_date'], format = '%Y-%m-%d')
```

```python
new_df[['children']]= new_df[['children']].astype('int64')
```

```python
new_df[['agent']] = new_df[['agent']].astype('int64')
```

```python
new_df[['company']] = new_df[['company']].astype('int64')
```

```python
new_df[['adr']] = new_df[['adr']].astype('int64')
```

```python
# Adding total staying days in hotels
new_df['total_stay'] = new_df['stays_in_weekend_nights']+new_df['stays_in_week_nights']

# Adding total people num as column, i.e. total people num = num of adults + children + babies
new_df['total_people'] = new_df['adults']+new_df['children']+new_df['babies']
```

**Question 10: what causes the cancelation of bookings of rooms by customers \ \ One question can arise that may be longer waiting period or longer lead time causes the cancellation of bookings, let us check that.**

```python
waiting_bookings = new_df[new_df['days_in_waiting_list'] !=0]  # Selecting bookings with non zero waiting time


fig, axes = plt.subplots(1, 2, figsize=(18, 8))
sns.kdeplot(ax=axes[0],x ='days_in_waiting_list', data = waiting_bookings, hue = 'is_canceled', palette = "viridis" )
sns.kdeplot(ax = axes[1], x = new_df['lead_time'], hue = new_df['is_canceled'], palette = "viridis")
plt.show()
```

We see that most of the bookings that are cancelled have waiting period of less 150 days but also most of bookings that are not cancelled also have waiting period less than 150 days. Hence this shows that waiting period has no effect on cancellation of bookings. \ \ Also, lead time has no affect on cancellation of bookings, as both curves of cancelation and not cncelation are similar for lead time too.
